## Análisis rápido de cuellos de botella

```python
resumen = MONITOR.resumen()
print(f'Cola maxima:    {resumen["cola_maxima"]} vehiculos')
print(f'Cola promedio:  {resumen["cola_promedio"]:.1f} vehiculos')
print(f'Total salidos:  {resumen["total_salidos"]:,} vehiculos')

# Hora del dia con mayor congestion
import numpy as np
hist = MONITOR._historia
cola_total = np.array(hist['cola_total'])
t_max = hist['t'][int(np.argmax(cola_total))]
hora_max = (t_max // 3600 + 7) % 24
min_max  = (t_max % 3600) // 60
print(f'\nPico de congestion: {hora_max:02d}:{min_max:02d} ({int(cola_total.max())} veh en cola)')
```
